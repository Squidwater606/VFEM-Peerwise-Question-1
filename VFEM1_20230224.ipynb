In this notebook, you will analyse a surface given by a position vector in the form r = x(u,v)i + y(u,v)j + z(u,v)k.
The surface, in this case, has a rather special property.

```python
# Importing libraries:

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Defining x, y & z as functions of u & v:

def x(u,v):
    return np.cos(u) * (1 + v/2*np.cos(u/2))

def y(u,v):
    return np.sin(u) * (1 + v/2*np.cos(u/2))

def z(u,v):
    return v/2*np.sin(u/2)
```

```python
# Creating a meshgrid of u, v (creates a grid of 'u' and 'v' values that can be used to evaluate the parametric equations of the surface):

u_vals = np.linspace(0, 2*np.pi, 100)
v_vals = np.linspace(-1, 1, 100)
u, v = np.meshgrid(u_vals, v_vals)
```

```python
# Coordinates for each u & v:

x_vals = x(u, v)
y_vals = y(u, v)
z_vals = z(u, v)
```

```python
# 3D plot of surface:

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_vals, y_vals, z_vals, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Setting the viewing angle:
ax.view_init(elev=30, azim=(90))

plt.show()
```

Now, find the vector normal to this surface:

```python
# Defining partial derivatives of x, y & z with respect to u & v:

def dx_du(u,v):
    return -((np.sin(u)*(1+(v/2)*np.cos(u/2))) + ((v/4)*np.cos(u)*np.sin(u/2)))
def dx_dv(u,v):
    return (1/2)*(np.cos(u/2))*(np.cos(u))
def dy_du(u,v):
    return ((np.cos(u))*(1+(v/2)*np.cos(u/2)) - (v/4)*(np.sin(u))*(np.sin(u/2)))
def dy_dv(u,v):
    return (1/2)*(np.cos(u/2))*(np.sin(u))
def dz_du(u,v):
    return (v/4)*(np.cos(u/2))
def dz_dv(u,v):
    return (1/2)*(np.sin(u/2))

# Defining vector normal to the surface (determinant of the Jacobian):

def normal(u,v):
    a = dx_du(u,v)
    b = dy_du(u,v)
    c = dz_du(u,v)
    d = dx_dv(u,v)
    e = dy_dv(u,v)
    f = dz_dv(u,v)
    
    n = np.array([b*f - c*e, c*d - a*f, a*e - b*d])
    return n / np.linalg.norm(n)
```
